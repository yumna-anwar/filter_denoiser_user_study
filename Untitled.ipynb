```python
import sqlite3

# Connect to the SQLite database
# Replace 'your_database.db' with the path to your database file
conn = sqlite3.connect('/Users/yumnaanwar/Desktop/filter_denoiser_user_study/server/data.db')
cursor = conn.cursor()

# Retrieve all rows from the table
# Replace 'your_table_name' with the name of the table you want to query
cursor.execute("SELECT * FROM UserStudiesPair WHERE UserId=17")
rows = cursor.fetchall()

ratings=[]
# Print all rows
for row in rows:
    print(row[2])
    print(row[7])
    print(row[3])
    ratings.append([row[2],row[7],row[3]])

# Close the connection
conn.close()
```

```python
import matplotlib.pyplot as plt
from collections import defaultdict, Counter
import re
data = ratings

# Extract the module and rating information
pair_rating_counts = defaultdict(Counter)
print(pair_rating_counts)
for row in data:
    mod1 = re.search(r'mod\d', row[0]).group()
    mod2 = re.search(r'mod\d', row[1]).group()
    
    if mod1=='mod1':
        mod1='h24_all'
    if mod1=='mod2':
        mod1='h24_one'
    if mod1=='mod3':
        mod1='h8_all'
    if mod1=='mod4':
        mod1='h8_one'
    if mod1=='mod5':
        mod1='Noisy'
    if mod1=='mod6':
        mod1='mmse'
        
    if mod2=='mod1':
        mod2='h24_all'
    if mod2=='mod2':
        mod2='h24_one'
    if mod2=='mod3':
        mod2='h8_all'
    if mod2=='mod4':
        mod2='h8_one'
    if mod2=='mod5':
        mod2='Noisy'
    if mod2=='mod6':
        mod2='mmse'
        
    rating = row[2]
    pair_rating_counts[(mod1, mod2)][rating] += 1

# Prepare data for plotting and plot each pair
ratings = ['A', 'B', 'SAME']

for pair, rating_counts in pair_rating_counts.items():
    mods = pair
    counts = [rating_counts[rating] for rating in ratings]
    
    plt.figure(figsize=(8, 6))
    plt.bar(ratings, counts, color=['blue', 'orange', 'green'])
    plt.xlabel('Ratings')
    plt.ylabel('Counts')
    plt.title(f'Counts of Ratings for Pair A={mods[0]} and B={mods[1]}')
    plt.show()
```
